# Load data

```python
df = load_parquet(filepath="33.1_CH-OE2_IRGA+LGR+METEO+MGMT_Level-1_eddypro_fluxnet_2021-2023.parquet")
df
```

# Testing mean settings to get mean over *preceding* records

```python
# Testing: closed='left' is the correct setting for preceding records
var = 'SWC_0.15_gfXG'
_df = df[[var]].copy()
_df['mean_left'] = _df[var].rolling(window=6, closed='left').mean()
_df['mean_right'] = _df[var].rolling(window=6, closed='right').mean()
display(_df[var].iloc[0:6].mean())
display(_df[var].iloc[1:7].mean())
display(_df.iloc[0:20])
```

# Calc 3-hour means from preceding records

```python
[print(c) for c in df.columns if "SWC_" in c];
[print(c) for c in df.columns if "TS_" in c];
[print(c) for c in df.columns if "PREC_" in c];
```

```python
var = ['SWC_0.05_gfXG', 'SWC_0.15_gfXG', 'SWC_0.3_gfXG', 'SWC_0.5_gfXG', 'TS_0.05_gfXG', 'TS_0.15_gfXG', 'TS_0.3_gfXG', 'TS_0.5_gfXG', 'PREC']

for v in var:    
    name_mean = f"{v}_MEAN3H"
    df[name_mean] = df[v].rolling(window=6, closed='left').mean()
```

```python
for v in var:
    print(f"\nResults for {v}: ")
    display(df[[v, f"{v}_MEAN3H"]].iloc[0:9])
```

# Save to file

```python
OUTNAME = "33.3_CH-OE2_IRGA+LGR+METEO+MGMT_Level-1_eddypro_fluxnet_2021-2023"
filepath = save_parquet(filename=OUTNAME, data=df)
df.to_csv(f"{OUTNAME}.csv")
```
