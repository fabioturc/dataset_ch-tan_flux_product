# Imports

```python
import importlib.metadata
import pandas as pd
import warnings
from datetime import datetime
from pathlib import Path

from diive.core.io.filereader import search_files, MultiDataFileReader
from diive.core.io.files import save_parquet, load_parquet
# pd.options.display.width = None
# pd.options.display.max_columns = None
pd.set_option('display.max_rows', 3000)
pd.set_option('display.max_columns', 3000)
warnings.filterwarnings(action='ignore', category=FutureWarning)
warnings.filterwarnings(action='ignore', category=UserWarning)
dt_string = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
version_diive = importlib.metadata.version("diive")
print(f"diive version: v{version_diive}")
```

# Load data: fluxes L1, meteo (10 variables) and management

```python
irga = load_parquet(filepath="31.1_CH-TAN_IRGA_Level-1_eddypro_fluxnet.parquet")
lgr = load_parquet(filepath="32.1_CH-TAN_LGR_Level-1_eddypro_fluxnet.parquet")
```

# Merge files

## Start with IRGA data

```python
df = irga.copy()
df
```

## Add *new* columns from QCL LGR data

```python
newcols = [c for c in lgr.columns if c not in df.columns]
df = pd.concat([df, lgr[newcols]], axis=1)
df
```

# Sort column order

```python
df = df.reindex(sorted(df.columns), axis=1)  # Sort column order
df
```

# Save to file

```python
OUTNAME = "33.1_CH-TAN_IRGA+LGR_Level-1_eddypro_fluxnet"
filepath = save_parquet(filename=OUTNAME, data=df)
df.to_csv(f"{OUTNAME}.csv")
```
