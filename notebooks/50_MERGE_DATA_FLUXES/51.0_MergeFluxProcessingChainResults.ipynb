# Imports

```python
from pathlib import Path
import pandas as pd
from diive.core.io.files import save_parquet, load_parquet
```

# Load L3.3 results for NEE, LE and H

```python
nee = load_parquet(filepath=r"../40_FLUX_PROCESSING_CHAIN/NEE/41.1_FluxProcessingChain_L3.2_NEE.parquet")
le = load_parquet(filepath=r"../40_FLUX_PROCESSING_CHAIN/LE/42.1_FluxProcessingChain_L3.2_LE.parquet")
h = load_parquet(filepath=r"../40_FLUX_PROCESSING_CHAIN/H/43.1_FluxProcessingChain_L3.2_H.parquet")
n2o = load_parquet(filepath=r"../40_FLUX_PROCESSING_CHAIN/FN2O/44.1_FluxProcessingChain_L3.2_FN2O.parquet")
ch4 = load_parquet(filepath=r"../40_FLUX_PROCESSING_CHAIN/FCH4/45.1_FluxProcessingChain_L3.2_FCH4.parquet")
```

# Load main data

```python
SOURCEDIR = r"../30_MERGE_DATA"
FILENAME = r"33.1_CH-TAN_IRGA+LGR+METEO+MGMT_Level-1_eddypro_fluxnet.parquet"
FILEPATH = Path(SOURCEDIR) / FILENAME
maindf = load_parquet(filepath=FILEPATH)
```

## Add results for NEE

```python
newcols = [c for c in nee.columns if c not in maindf]
print("NEW VARIABLES FROM FLUX PROCESSING CHAIN:")
[print(f"++ {c}") for c in newcols]
maindf = pd.concat([maindf, nee[newcols]], axis=1)
maindf
```

## Add results for H

```python
newcols = [c for c in h.columns if c not in maindf]
print("NEW VARIABLES FROM FLUX PROCESSING CHAIN:")
[print(f"++ {c}") for c in newcols]
maindf = pd.concat([maindf, h[newcols]], axis=1)
maindf
```

## Add results for N2O

# Export 

## Export all data

```python
filename = "51.1_FLUXES_METEO_MGMT_L3.2_NEE_LE_H_FN2O_FCH4"
maindf.to_csv(f"{filename}.csv", index=True)
save_parquet(data=maindf, filename=filename)
```

## Export subset for REddyProc

```python
from diive.core.times.times import insert_timestamp
export_REddyProc = maindf.copy()
```

```python
exportcols = ['NEE_L3.1_L3.2_QCF', 'LE_L3.1_L3.2_QCF', 'H_L3.1_L3.2_QCF', 'FN2O_L3.1_L3.2_QCF', 'FCH4_L3.1_L3.2_QCF',
              'USTAR',
              'sw_in', 'ta', 'vpd']
export_REddyProc = export_REddyProc[exportcols].copy()
locs = (export_REddyProc.index.year >= 2023) & (export_REddyProc.index.year <= 2025)
export_REddyProc = export_REddyProc[locs].copy()
export_REddyProc = insert_timestamp(data=export_REddyProc, convention="end", set_as_index=True)
export_REddyProc
```

```python
filename = "51.2_FluxProcessingChain_L3.2_subset-forREddyProc"
export_REddyProc.to_csv(f"{filename}.csv", index=True)
save_parquet(data=export_REddyProc, filename=filename)
```
